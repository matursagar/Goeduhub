1.	Load the necessary package for plotting using pyplot from matplotlib. \\
Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement. \\
Days=[1,2,3,4,5,6,7,8] \\
Speed=[60,62,61,58,56,57,46,63]


```python
#Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
Days = [1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]
```

```python
plt.plot(Days, Speed)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car Speed")
plt.show()
```

2. Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.

```python
plt.plot(Days, Speed, color='green', linestyle='dotted', marker='+', markersize='10', markerfacecolor='r')
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car Speed")
plt.show()
```

3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days. \\
days=[1,2,3,4,5,6,7,8] \\
max_speed=[80,91,92,88,77,79,76,75] \\
min_speed=[42,43,40,42,33,36,34,35] \\

```python
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days, max_speed, 'g', label='max speed', linestyle='--', marker='+', markersize='10')
plt.plot(days, min_speed, 'r', label='min speed', linestyle='dashdot', marker='+', markersize='10')
plt.plot(days, avg_speed, 'b', label='avg speed', linestyle='dotted', marker='+', markersize='10')
plt.legend()
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car Speed")
plt.show()
```

4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
import math
X = np.arange(0, math.pi*2, 0.05)
  
# Assign variables to the y axis part of the curve
y = np.sin(X)
z = np.cos(X)
plt.plot(X, y, color='r', label='sin')
plt.plot(X, z, color='b', label='cos')

plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine and Cosine curves")
plt.legend()
plt.show()
```

5. Plot Simple bar chart showing popularity of Programming Languages. \\
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript'] \\
Popularity = [56, 39, 34, 34, 29] \\
Security = [44 ,36 ,55, 50, 42] \\
Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.


```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
plt.bar(Languages, Popularity)
plt.title("Popularity of Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
x_axis =np.arange(len(Languages))
plt.bar(x_axis-0.2, Popularity, label="Popularity", color='r', width=0.4)
plt.bar(x_axis+0.2, Security, label="Security", color='g', width=0.4)
plt.title("Popularity of Programming Languages")
plt.xticks(x_axis, Languages)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.legend()
plt.show()
```

7.  Titanic Data Set Download Data \\
Load the data file \\
(i) Create a pie chart presenting the male/female proportion \\
(ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender


```python
df = pd.read_csv("/content/titanic_original_assignment6.csv")
```

```python
df.columns
```

```python
a = df.groupby('sex')['sex'].count()
print(a)
```

```python
#Create a Pie plot
labels = ['Female', 'Male']
colors = ['tab:green', 'tab:red']
fig, ax = plt.subplots()
ax.pie(a, labels = labels,
          colors = colors,
          autopct='%.0f%%',
          startangle = 180)

ax.set_title('Male/Female Proportion')
plt.show()
```

```python
#Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
plt.figure(figsize=(9,6))
ax = plt.subplot()
ax.scatter(df[df['sex']=='male']['age'],df[df['sex']=='male']['fare'],c='green',s=40)
ax.scatter(df[df['sex']=='female']['age'],df[df['sex']=='female']['fare'],c='red',s=40)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.legend(('Male','Female'),scatterpoints=1,loc='upper right',fontsize=15,)
plt.show()
```
